```python
import numpy as np
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
import numpy.random as random
%matplotlib inline
%precision 3
np.random.seed(0)
```

```python
import requests,zipfile
from io import StringIO
import io

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00356/student.zip"
r = requests.get(url, stream=True)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall()
```

```python

from scipy import stats

student_data_math = pd.read_csv("student-mat.csv", sep=';')
student_data_por = pd.read_csv("student-por.csv", sep=';')
student_data_merge = pd.merge(student_data_math, student_data_por, on=["school","sex","age","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"],suffixes=('_math','_por'))

```

#総合問題

```python
print("数学の欠席者平均:",student_data_merge.absences_math.mean())
print("ポルトガル語の欠席者平均:",student_data_merge.absences_por.mean())

t,p = stats.ttest_rel(student_data_merge.absences_math,student_data_merge.absences_por)
print("p値:",p)
```

```python
print("数学の勉強時間平均:",student_data_merge.studytime_math.mean())
print("ポルトガル語の勉強時間平均:",student_data_merge.studytime_por.mean())

t,p = stats.ttest_rel(student_data_merge.studytime_math,student_data_merge.studytime_por)
print("p値:",p)
```
