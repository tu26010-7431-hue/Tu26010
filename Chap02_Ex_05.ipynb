```python
import numpy as np
import scipy.linalg as linalg

a = np.array([[1,2,3],[1,3,2],[3,1,2]])

print("逆行列")
print(linalg.inv(a))

eig_value, eig_vector = linalg.eig(a)
print("固有値")
print(eig_value)
print("固有ベクトル")
print(eig_vector)
```
