```python
import numpy as np
import numpy.random as random
import scipy as sp
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
%precision 3
```

```python
from scipy.optimize import minimize

def func(x):
  return x**2+1
def cons(x):
  return(x+1)
cons = ({'type':'ineq','fun':cons})
x = 0
result = minimize(func,x0=x,constraints=cons,method='SLSQP')
print(result)
```

```python
print('Y:',result.fun)
print('X:',result.x)
```
