```python
import math
import numpy as np
import matplotlib.pyplot as plt

def uni_hist(N):
  x = np.random.uniform(0.0,1.0,N)
  y = np.random.uniform(0.0,1.0,N)
  plt.subplot(2,1,1)
  plt.hist(x)
  plt.title("No1:histogram")

  plt.subplot(2,1,2)
  plt.hist(y)
  plt.title("No2:histogram")
  plt.grid(True)

  plt.tight_layout()
uni_hist(1000)
```
