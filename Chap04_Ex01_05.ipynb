```python
import numpy as np
import scipy as sp
import pandas as pd
from pandas import Series, DataFrame
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib as mpl
%matplotlib inline
%precision 3
np.random.seed(0)
```

```python
#　練習問題4-2
coin_data = np.array([0,1])
N = 1000
count_coin = np.random.choice(coin_data,N)
for i in range (0,2):
  p = len(count_coin[count_coin==i]) / N
  print(i,"が出る確率",p)
```
