```python
#　練習問題4-13
from scipy import stats

student_data_math = pd.read_csv("student-mat.csv", sep=';')
student_data_por = pd.read_csv("student-por.csv", sep=';')
student_data_merge = pd.merge(student_data_math, student_data_por, on=["school","sex","address","famsize","Pstatus","Medu","Fedu","Mjob","Fjob","reason","nursery","internet"],how="inner",suffixes=('_math','_por'))

print("G2数学平均:",student_data_merge.G2_math.mean())
print("G2ポルトガル語平均:",student_data_merge.G2_por.mean())

t,p = stats.ttest_rel(student_data_merge.G2_math,student_data_merge.G2_por)
print("t値:",t)
print("p値:",p)

print("G3数学平均:",student_data_merge.G3_math.mean())
print("G3ポルトガル語平均:",student_data_merge.G3_por.mean())

t,p = stats.ttest_rel(student_data_merge.G3_math,student_data_merge.G3_por)
print("t値:",t)
print("p値:",p)
```
